```python
#@title Setup pipeline
#@title Install dependencies
!pip install torch==2.0.0 torchvision==0.15.0 -f https://download.pytorch.org/whl/cu118/torch_stable.html git+https://github.com/huggingface/diffusers transformers accelerate imageio[ffmpeg]
import torch
from diffusers import DiffusionPipeline, DPMSolverMultistepScheduler
from diffusers.utils import export_to_video
from IPython.display import HTML
from base64 import b64encode

pipe = DiffusionPipeline.from_pretrained("damo-vilab/text-to-video-ms-1.7b", torch_dtype=torch.float16, variant="fp16")
pipe.scheduler = DPMSolverMultistepScheduler.from_config(pipe.scheduler.config)
pipe.enable_model_cpu_offload()
pipe.enable_vae_slicing()
```

```python
import io
import imageio
import anvil.media
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML
import io
import base64


@anvil.server.callable
def Generate_video(Text_box, Time_box):
    # Your video generation logic here
    prompt =  Text_box
    video_duration_seconds =  Time_box
    num_frames = video_duration_seconds * 10
    video_frames = pipe(prompt, negative_prompt="low quality", num_inference_steps=25, num_frames=num_frames).frames
    video_frames = video_frames[0]
    # video_path = export_to_video(video_frames)
    # Assuming video_bytes is your generated video in bytes
    # Ensure you have video_frames and video_bytes correctly generated

    # Convert video_frames to uint8 type
    video_frames = (video_frames * 255).astype('uint8')

    with io.BytesIO() as video_buffer:
        with imageio.get_writer(video_buffer, format='mp4', fps=10) as writer:
            for frame in video_frames:
                writer.append_data(frame)
        video_bytes = video_buffer.getvalue()
    video_base64 = base64.b64encode(video_bytes).decode('utf-8')

    # Return the video bytes
    return video_base64
```
